```python
# Step 0: Load libraries
import pandas as pd
import numpy as np

# Step 1: Load your dataset (CSV)
from google.colab import files
uploaded = files.upload()  # Upload your CSV file here
csv_file = list(uploaded.keys())[0]

# Load into DataFrame
df = pd.read_csv(csv_file)

# Step 2: Creating a DataFrame
dataframe = pd.DataFrame(df)
print("DataFrame created. First 2 rows:\n")
print(dataframe.head(2))

# Step 3: Describe the dataset
print("\nDimensions:", df.shape)
print("\nStatistics of numerical columns:\n")
print(df.describe())

# Step 4: Navigating DataFrames
print("\nSelect first row:\n", df.iloc[0])
print("\nSelect rows 2-4:\n", df.iloc[1:4])
print("\nAll rows up to 4th row:\n", df.iloc[:4])

# Step 5: Set index to a unique column (if applicable)
# Assuming 'id' or 'name' exists, else skip
if 'id' in df.columns:
    df.set_index('id', inplace=True)
    print("\nRow with index=1:\n", df.loc[1])

# Step 6: Selecting rows based on conditionals
print("\nTop 2 rows where gender column is 'Female':\n")
if 'gender' in df.columns:
    print(df[df['gender'] == 'Female'].head(2))

# Multiple conditions example
if 'age' in df.columns and 'gender' in df.columns:
    print("\nRows where gender is 'Female' and age >= 65:\n")
    print(df[(df['gender'] == 'Female') & (df['age'] >= 65)])

# Step 7: Replacing values
if 'gender' in df.columns:
    df['gender'] = df['gender'].replace(['female', 'male'], ['Woman', 'Man'])
    print("\nFirst 5 values after replacing gender:\n", df['gender'].head(5))

# Step 8: Renaming columns
df.rename(columns={'gender': 'Gender', 'age': 'Age'}, inplace=True)
print("\nColumns after renaming:\n", df.head(2))

# Step 9: Finding Min, Max, Mean, Sum, Count
if 'Age' in df.columns:
    print("\nMaximum Age:", df['Age'].max())
    print("Minimum Age:", df['Age'].min())
    print("Mean Age:", df['Age'].mean())
    print("Sum of Age:", df['Age'].sum())
    print("Count of Age:", df['Age'].count())

# Step 10: Variance, Std, Kurtosis, Skewness
print("\nVariance:\n", df.var())
print("\nStandard Deviation:\n", df.std())
print("\nKurtosis:\n", df.kurt())
print("\nSkewness:\n", df.skew())

# Step 11: Finding Unique Values and Counts
if 'Gender' in df.columns:
    print("\nUnique genders:\n", df['Gender'].unique())
    print("\nGender counts:\n", df['Gender'].value_counts())

# Step 12: Handling missing values
if 'Age' in df.columns:
    print("\nRows with missing Age values:\n", df[df['Age'].isnull()].head(2))

# Step 13: Deleting a column
if 'Age' in df.columns:
    df_drop_col = df.drop('Age', axis=1)
    print("\nAfter deleting 'Age' column:\n", df_drop_col.head(2))

# Step 14: Deleting a row
# Example: delete rows where Gender == 'Man'
df_drop_row = df[df['Gender'] != 'Man']
print("\nAfter deleting rows where Gender is 'Man':\n", df_drop_row.head(2))

# Step 15: Grouping rows
if 'Gender' in df.columns and 'stroke' in df.columns:
    print("\nGroup by Gender, mean values:\n", df.groupby('Gender').mean())
    print("\nCount of stroke cases by Gender:\n", df.groupby('Gender')['stroke'].count())
    print("\nCount of stroke cases:\n", df.groupby('stroke')['Gender'].count())
    print("\nMean Age by Gender & Stroke:\n", df.groupby(['Gender', 'stroke'])['Age'].mean())

# Step 16: Looping over a column
print("\nUppercase first 2 names:")
if 'name' in df.columns:
    for name in df['name'][:2]:
        print(name.upper())

# Step 17: Applying a function over a column
def uppercase(x):
    return str(x).upper()

if 'name' in df.columns:
    df['name_upper'] = df['name'].apply(uppercase)
    print("\nFirst 2 names after apply function:\n", df['name_upper'].head(2))

# Step 18: Applying function to groups
if 'Gender' in df.columns:
    print("\nCount of each column grouped by Gender:\n", df.groupby('Gender').apply(lambda x: x.count()))

# Step 19: Concatenating DataFrames (example)
# Split df into two halves and concatenate
df1 = df.head(len(df)//2)
df2 = df.tail(len(df)//2)
df_concat = pd.concat([df1, df2])
print("\nConcatenated DataFrame shape:", df_concat.shape)
```
