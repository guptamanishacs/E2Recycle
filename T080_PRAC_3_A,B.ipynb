```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler, RobustScaler

# Load file uploaded in Colab
df = pd.read_csv("Plant_1_Generation_Data.csv")   # change if name is different

print("Numeric columns:")
print(df.select_dtypes(include=['float64', 'int64']).columns)

# Select the column to scale
column = "DC_POWER"

feature = df[[column]].values

print("\nOriginal Feature (first 10 values):")
print(feature[:10])

# Standard Scaler
scaler = StandardScaler()
standardized = scaler.fit_transform(feature)

print("\nStandardized Values (first 10):")
print(standardized[:10])

print("\nMean:", round(standardized.mean(), 4))
print("Std Dev:", round(standardized.std(), 4))

# Robust Scaler
robust_scaler = RobustScaler()
robust_scaled = robust_scaler.fit_transform(feature)

print("\nRobust Scaled Values (first 10):")
print(robust_scaled[:10])
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import Normalizer

# ---------------------------------------------
# STEP 1 — Load Uploaded Dataset
# ---------------------------------------------
df = pd.read_csv("Plant_1_Generation_Data.csv")   # use exact uploaded filename

# Select 2 numeric columns for normalization, ensuring non-zero values for demonstration
# Filter for rows where at least one of the power values is greater than 0
features = df[(df["DC_POWER"] > 0) | (df["AC_POWER"] > 0)][["DC_POWER", "AC_POWER"]].head(5).values

# If no non-zero values are found in the head(5), try a larger sample or fall back to original head(5)
if features.shape[0] == 0:
    print("Warning: No non-zero power values found in the first few rows. Using original head(5).")
    features = df[["DC_POWER", "AC_POWER"]].head(5).values

print("Original Feature Matrix (first 5 rows with non-zero power):")
print(features)

# ---------------------------------------------
# STEP 2 — L2 Normalization (unit length = 1)
# ---------------------------------------------
normalizer_l2 = Normalizer(norm="l2")
features_l2_norm = normalizer_l2.transform(features)

print("\nL2 Normalized Matrix:")
print(features_l2_norm)

# ---------------------------------------------
# STEP 3 — L1 Normalization (sum of row = 1)
# ---------------------------------------------
normalizer_l1 = Normalizer(norm="l1")
features_l1_norm = normalizer_l1.transform(features)

print("\nL1 Normalized Matrix:")
print(features_l1_norm)

# Confirm L1 normalization (only for rows that are not all zeros)
if features_l1_norm.shape[0] > 0 and not np.all(features_l1_norm[0] == 0):
    print("\nSum of first row (should be 1):")
    print(features_l1_norm[0,0] + features_l1_norm[0,1])
else:
    print("\nCannot confirm L1 sum to 1 as the selected features are all zeros.")
```

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

df = pd.read_csv("Plant_1_Generation_Data.csv")   # use exact uploaded filename

# Select the first 1000 rows (or more, if needed) to ensure data variation
features = df[["DC_POWER", "AC_POWER"]].dropna().head(1000)

print("Selected Features (first 5 rows):")
print(features.head())

# Convert to array
X = features.values

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Initialize KMeans with 3 clusters
clusterer = KMeans(n_clusters=3, random_state=0, n_init='auto') # Added n_init='auto' to suppress future warning

# Fit
clusterer.fit(X_scaled)

# Predict cluster group
features["group"] = clusterer.predict(X_scaled)

# ---------------------------------------------
# STEP 5 — Show final output
# ---------------------------------------------
print("\nFinal Clustered Output:")
print(features.head())
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer

# STEP 1 — Load Dataset
df = pd.read_csv("Plant_1_Generation_Data.csv")   # use exact uploaded filename

features = df[["DC_POWER", "AC_POWER"]].dropna().head(1000).values

print("Original Features (first 5 rows):")
print(features[:5])

scaler = StandardScaler()
standardized_features = scaler.fit_transform(features)

# Store true value BEFORE inserting NaN
true_value = standardized_features[0, 0]

# Intentionally introduce a NaN for demonstration purposes
standardized_features[0, 0] = np.nan

print("\nTrue Value before missing:", true_value)

# Show the column mean with the introduced NaN, for clarity
print("Mean of column 0 with NaN (before imputation):", np.nanmean(standardized_features[:, 0]))

# STEP 3 — Create Imputer (Mean Strategy)
mean_imputer = SimpleImputer(strategy="mean")

# Impute missing values
features_mean_imputed = mean_imputer.fit_transform(standardized_features)

# STEP 4 — Show results
print("\nImputed Value:", features_mean_imputed[0, 0])
```

```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction import DictVectorizer
from sklearn.impute import SimpleImputer

df = pd.read_csv("/content/Plant_1_Generation_Data.csv")

df["DC_CATEGORY"] = pd.cut(df["DC_POWER"],
                           bins=[-1, 500, 2000, 7000],
                           labels=["Low", "Medium", "High"])

df["DC_CATEGORY"] = df["DC_CATEGORY"].astype(str)
scale_mapper = {"Low": 1, "Medium": 2, "High": 3}
ordinal_encoded = df["DC_CATEGORY"].map(scale_mapper)
print(ordinal_encoded.head())

dict_data = df[["DC_POWER", "AC_POWER"]].head(4).to_dict(orient="records")
vec = DictVectorizer(sparse=False)
dict_encoded = vec.fit_transform(dict_data)
print(dict_encoded)
print(vec.get_feature_names_out())

df_missing = df[["AC_POWER", "DC_POWER"]].copy()
df_missing.loc[0, "AC_POWER"] = np.nan

imp = SimpleImputer(strategy="mean")
df_filled = imp.fit_transform(df_missing)

print(df_filled[:5])
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer

df = pd.read_csv("Plant_1_Generation_Data.csv")

features = df[["DC_POWER", "AC_POWER"]].values

scaler = StandardScaler()
standardized_features = scaler.fit_transform(features)

true_value = standardized_features[0, 0]
standardized_features[0, 0] = np.nan

mean_imputer = SimpleImputer(strategy="mean")
features_mean_imputed = mean_imputer.fit_transform(standardized_features)

print("True Value: {}".format(true_value))
print("Imputed Value: {}".format(features_mean_imputed[0, 0]))
```
